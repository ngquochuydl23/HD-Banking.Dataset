```python
import random
import pandas as pd
import numpy as np
from faker import Faker
from vn_fullname_generator import generator
```

```python
danhsach_nganhang = pd.read_csv('banking_resource_production.csv')
danhsach_nganhang.head(5)
```

```python

fake = Faker('vi_VN')

banks = danhsach_nganhang['ShortName'].to_numpy()
bins = danhsach_nganhang['Bin'].to_numpy()
account_status = ['Active', 'Inactive', 'Closed', 'Frozen']
gender = [0, 1]
def generate_fake_bank_data(num_records=1000000):
    data = []
    for _ in range(num_records):

        bank = random.choice(banks)
        bank_index = np.where(banks == bank)
        bin = bins[bank_index][0]

        account_number = fake.random_number(digits=11, fix_len=True)
        account_holder = generator.generate(random.choice(gender))
        bank = bank
        balance = random.randint(0, 100000000)
        open_date = fake.date_between(start_date='-5y', end_date='today')
        status = random.choices(account_status, weights=[0.6, 0.2, 0.1, 0.1], k=1)[0]
        cccd = str(fake.random_int(min=100000000000, max=999999999999, step=1))  # Số CCCD 12 chữ số
        cccd = cccd.zfill(12)
        data.append([account_number, account_holder, cccd, bank, bin, balance, open_date, status])

    columns = ['AccountNo', 'OwnerName', 'IdCardNo', 'BankName', 'Bin', 'Balance', 'OpenedAt', 'Status']
    return pd.DataFrame(data, columns=columns)

df = generate_fake_bank_data()
df.drop_duplicates(subset='AccountNo', inplace=True)
df.to_csv('fake_bank_accounts.csv', index=False, encoding='utf-8')

print("Dataset đã được tạo và lưu vào file 'fake_bank_accounts.csv'")
```
